```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("masoudnickparvar/brain-tumor-mri-dataset")

print("Path to dataset files:", path)
```

```python
import os
import cv2
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# Rutas a las carpetas de entrenamiento y prueba
train_dir = os.path.join(path, 'Training')
test_dir = os.path.join(path, 'Testing')
```

```python
# Clases de tumores
classes = ['glioma', 'meningioma', 'notumor', 'pituitary']
```

```python

# Función para cargar imágenes y etiquetas
def load_data(directory):
    images = []
    labels = []
    for class_name in classes:
        class_path = os.path.join(directory, class_name)
        for img_name in os.listdir(class_path):
            img_path = os.path.join(class_path, img_name)
            try:
                img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) # Cargar en escala de grises
                if img is not None:
                    # El dataset menciona que las imágenes tienen diferentes tamaños.
                    # Redimensionar a un tamaño común (ej. 128x128) para uniformidad.
                    img = cv2.resize(img, (128, 128))
                    images.append(img)
                    labels.append(class_name)
                else:
                    print(f"Advertencia: No se pudo cargar la imagen {img_path}")
            except Exception as e:
                print(f"Error al procesar la imagen {img_path}: {e}")
    return np.array(images), np.array(labels)
```

```python
print("Cargando datos de entrenamiento...")
X_train, y_train = load_data(train_dir)
print("Cargando datos de prueba...")
X_test, y_test = load_data(test_dir)

print(f"Forma de los datos de entrenamiento: {X_train.shape}, Etiquetas: {y_train.shape}")
print(f"Forma de los datos de prueba: {X_test.shape}, Etiquetas: {y_test.shape}")
```

```python
# Codificar etiquetas categóricas a numéricas
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Convertir a one-hot encoding para el entrenamiento de la CNN
y_train_categorical = to_categorical(y_train_encoded, num_classes=len(classes))
y_test_categorical = to_categorical(y_test_encoded, num_classes=len(classes))
```

```python
print(f"Ejemplo de etiquetas codificadas: {y_train_encoded[:5]}")
print(f"Ejemplo de etiquetas one-hot: {y_train_categorical[:5]}")
print(f"Clases mapeadas: {list(label_encoder.classes_)}")
```

```python
# Normalizar los valores de píxeles a un rango de 0 a 1
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
```

```python
# Añadir una dimensión de canal para las imágenes en escala de grises (requerido por Keras/TensorFlow)
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)
```

```python
print(f"Forma final de los datos de entrenamiento (normalizados y con canal): {X_train.shape}")
print(f"Forma final de los datos de prueba (normalizados y con canal): {X_test.shape}")
```

```python
# Visualizar algunas imágenes de ejemplo
plt.figure(figsize=(10, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(X_train[i].squeeze(), cmap='gray')
    plt.title(f'Clase: {y_train[i]}')
    plt.axis('off')
plt.tight_layout()
plt.show()
```

#Paso 2: Aumento de Datos (Data Augmentation)

El aumento de datos es una técnica crucial para mejorar la robustez y la capacidad de generalización de los modelos de aprendizaje profundo, especialmente cuando el tamaño del dataset es limitado. Al aplicar transformaciones aleatorias a las imágenes de entrenamiento, se crea una mayor diversidad de datos, lo que ayuda al modelo a ser menos sensible a variaciones en la posición, orientación o iluminación de las imágenes.

Para este dataset, que consiste en imágenes 2D, podemos aplicar transformaciones como rotaciones, volteos horizontales/verticales, zooms, cambios de brillo y desplazamiento. TensorFlow/Keras proporciona herramientas convenientes para esto a través de ImageDataGenerator o capas de pre-procesamiento de Keras.



```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Crear un generador de aumento de datos
datagen = ImageDataGenerator(
    rotation_range=20,        # Rotación aleatoria de hasta 20 grados
    width_shift_range=0.1,    # Desplazamiento horizontal aleatorio (fracción del ancho total)
    height_shift_range=0.1,   # Desplazamiento vertical aleatorio (fracción del alto total)
    shear_range=0.1,          # Cizallamiento aleatorio
    zoom_range=0.1,           # Zoom aleatorio
    horizontal_flip=True,     # Volteo horizontal aleatorio
    vertical_flip=True,       # Volteo vertical aleatorio
    fill_mode='nearest'       # Estrategia para rellenar píxeles nuevos creados por transformaciones
)
```

```python
for X_batch, y_batch in datagen.flow(X_train, y_train_categorical, batch_size=9):
     plt.figure(figsize=(10, 10))
     for i in range(9):
         plt.subplot(3, 3, i + 1)
         plt.imshow(X_batch[i].squeeze(), cmap='gray')
         plt.title(f'Clase: {label_encoder.inverse_transform([np.argmax(y_batch[i])])[0]}')
         plt.axis('off')
     plt.tight_layout()
     plt.show()
```

# Paso 3: Selección y Diseño del Modelo

Para la clasificación de imágenes, las Redes Neuronales Convolucionales (CNNs) son la elección estándar. Dada la naturaleza de las imágenes médicas y el tamaño del dataset, podemos considerar dos enfoques:

1.
Construir una CNN desde cero: Un modelo más simple que se entrena completamente con nuestros datos.

2.
Usar Transfer Learning: Utilizar un modelo pre-entrenado en un dataset grande (como ImageNet) y adaptarlo a nuestra tarea. Esto es muy efectivo cuando se tienen datasets de tamaño moderado, ya que los modelos pre-entrenados ya han aprendido características visuales generales.

Optaremos por el enfoque de Transfer Learning por su eficacia probada en tareas de clasificación de imágenes médicas. Utilizaremos un modelo como VGG16, ResNet50 o MobileNetV2, y lo ajustaremos para nuestras 4 clases.



##3.1. Modelo Base (VGG16 con Transfer Learning)

Cargaremos el modelo VGG16 pre-entrenado en ImageNet, excluyendo las capas superiores (clasificador) y añadiendo nuestras propias capas para la clasificación de tumores cerebrales.



```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Input, concatenate # Import Input and concatenate

# Definir las dimensiones de entrada de la imagen (altura, ancho, canales)
IMG_HEIGHT = 128
IMG_WIDTH = 128
IMG_CHANNELS = 1 # Imágenes en escala de grises
input_shape = (IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS)
num_classes = len(classes)
```

```python
# Cargar el modelo base VGG16 pre-entrenado en ImageNet
# include_top=False: No incluir las capas de clasificación finales de VGG16
# weights=‘imagenet’: Usar pesos pre-entrenados de ImageNet
# input_shape: Definir la forma de entrada de nuestras imágenes

# Nota: VGG16 espera 3 canales de entrada. Para imágenes en escala de grises (1 canal),
# necesitamos replicar el canal para que coincida con la expectativa del modelo.
# Esto se hará en la capa de entrada o durante el pre-procesamiento.

# Para VGG16, replicamos el canal de gris a 3 canales RGB
input_tensor = Input(shape=input_shape)
x = concatenate([input_tensor, input_tensor, input_tensor], axis=-1) # Replicar canal para VGG16
base_model = VGG16(weights='imagenet', include_top=False, input_tensor=x)
```

```python
# Congelar las capas del modelo base para no re-entrenarlas inicialmente
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Añadir capas de clasificación personalizadas
x = base_model.output
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(num_classes, activation='softmax')(x)
```

```python
# Crear el modelo final
model = Model(inputs=base_model.input, outputs=predictions)

# Compilar el modelo
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

#Paso 4: Entrenamiento del Modelo

El entrenamiento del modelo implica alimentar los datos de entrenamiento a la red, ajustar sus pesos para minimizar la función de pérdida y monitorear su rendimiento en el conjunto de validación. Utilizaremos el ImageDataGenerator creado anteriormente para alimentar los datos aumentados al modelo.



```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau

# Configurar callbacks para el entrenamiento
# ModelCheckpoint: Guarda el mejor modelo basado en la precisión de validación
checkpoint = ModelCheckpoint(
    'best_brain_tumor_classifier.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)
```

```python
# EarlyStopping: Detiene el entrenamiento si la precisión de validación no mejora después de un número de épocas
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    mode='max',
    verbose=1,
    restore_best_weights=True
)
```

```python
# ReduceLROnPlateau: Reduce la tasa de aprendizaje cuando la métrica de monitoreo deja de mejorar
reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.1,
    patience=5,
    min_lr=1e-7,
    mode='max',
    verbose=1
)
```

```python
callbacks_list = [checkpoint, early_stopping, reduce_lr]

# Entrenar el modelo

epochs = 50
batch_size = 32
```

```python
print("Iniciando entrenamiento del modelo...")
history = model.fit(
    datagen.flow(X_train, y_train_categorical, batch_size=batch_size),
    steps_per_epoch=len(X_train) // batch_size,
    epochs=epochs,
    validation_data=(X_test, y_test_categorical), # Usamos X_test como conjunto de validación aquí
    callbacks=callbacks_list
)

print("Entrenamiento completado.")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Precisión de Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión de Validación')
plt.title('Precisión del Modelo')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de Validación')
plt.title('Pérdida del Modelo')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()

plt.tight_layout()
plt.show()
```

#Paso 5: Evaluación del Modelo

Una vez que el modelo ha sido entrenado, es crucial evaluar su rendimiento en el conjunto de prueba para obtener una estimación imparcial de su capacidad de generalización. Además de la precisión, examinaremos otras métricas como la matriz de confusión y el reporte de clasificación.

```python
import os
import cv2
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import kagglehub
```

```python
# Get the path to the dataset files
path = kagglehub.dataset_download("masoudnickparvar/brain-tumor-mri-dataset")

# Define the paths for training and testing data
training = os.path.join(path, "Training")
testing = os.path.join(path, "Testing")

# Define the labels
etiquetas = ['glioma','meningioma','pituitary','notumor']

# ------------------------------
# Función para cargar imágenes
# ------------------------------
def imagenes(path, etiquetas):
    lista_imagenes = []
    labels = []
    for label in etiquetas:
        for img in os.listdir(os.path.join(path, label)):
            img = cv2.imread(os.path.join(path, label, img), cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (128, 128))
            lista_imagenes.append(img)
            labels.append(label)
    return np.array(lista_imagenes), np.array(labels)

# ------------------------------
# Cargar datasets
# ------------------------------
X_train, Y_train = imagenes(training, etiquetas)
X_test, Y_test = imagenes(testing, etiquetas)

# Normalizar (0-1)
X_train = X_train.astype("float32") / 255.0
X_test = X_test.astype("float32") / 255.0

# Añadir canal (para CNN en escala de grises)
X_train = np.expand_dims(X_train, axis=-1)
X_test = np.expand_dims(X_test, axis=-1)

# ------------------------------
# Codificar etiquetas
# ------------------------------
le = LabelEncoder()
Y_train_encoded = le.fit_transform(Y_train)
Y_test_encoded = le.transform(Y_test)

Y_train_categorical = to_categorical(Y_train_encoded, num_classes=len(etiquetas))
Y_test_categorical = to_categorical(Y_test_encoded, num_classes=len(etiquetas))

# ------------------------------
# Definir modelo
# ------------------------------
def cnn_model_dropout(input_shape=(128, 128, 1)):
    model = Sequential()
    model.add(Conv2D(32, (3,3), padding='same', activation='relu', input_shape=input_shape))
    model.add(Conv2D(32, (3,3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))

    model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
    model.add(Conv2D(64, (3,3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))

    model.add(Conv2D(128, (3,3), padding='same', activation='relu'))
    model.add(Conv2D(128, (3,3), padding='same', activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    #model.add(Dropout(0.5))
    model.add(Dense(len(etiquetas), activation='softmax'))

    return model

# ------------------------------
# Compilar modelo
# ------------------------------
model = cnn_model_dropout(input_shape=(128, 128, 1))
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
checkpoint = ModelCheckpoint(
    'best_brain_tumor_classifier.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)
```

```python
# EarlyStopping: Detiene el entrenamiento si la precisión de validación no mejora después de un número de épocas
early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    mode='max',
    verbose=1,
    restore_best_weights=True
)
```

```python
# ReduceLROnPlateau: Reduce la tasa de aprendizaje cuando la métrica de monitoreo deja de mejorar
reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.1,
    patience=5,
    min_lr=1e-7,
    mode='max',
    verbose=1
)
```

```python
callbacks_list = [checkpoint, early_stopping, reduce_lr]

# Entrenar el modelo

epochs = 50
batch_size = 32
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Crear un generador de aumento de datos
datagen = ImageDataGenerator(
    rotation_range=20,        # Rotación aleatoria de hasta 20 grados
    width_shift_range=0.1,    # Desplazamiento horizontal aleatorio (fracción del ancho total)
    height_shift_range=0.1,   # Desplazamiento vertical aleatorio (fracción del alto total)
    shear_range=0.1,          # Cizallamiento aleatorio
    zoom_range=0.1,           # Zoom aleatorio
    horizontal_flip=True,     # Volteo horizontal aleatorio
    vertical_flip=True,       # Volteo vertical aleatorio
    fill_mode='nearest'       # Estrategia para rellenar píxeles nuevos creados por transformaciones
)
```

```python
print("Iniciando entrenamiento del modelo...")
history = model.fit(
    datagen.flow(X_train, Y_train_categorical, batch_size=batch_size),
    steps_per_epoch=len(X_train) // batch_size,
    epochs=epochs,
    validation_data=(X_test, Y_test_categorical), # Usamos X_test como conjunto de validación aquí
    callbacks=callbacks_list
)

print("Entrenamiento completado.")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Precisión de Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Precisión de Validación')
plt.title('Precisión del Modelo')
plt.xlabel('Época')
plt.ylabel('Precisión')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Pérdida de Entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de Validación')
plt.title('Pérdida del Modelo')
plt.xlabel('Época')
plt.ylabel('Pérdida')
plt.legend()

plt.tight_layout()
plt.show()
```
